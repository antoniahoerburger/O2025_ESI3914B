# Description
You’re given listening data from a music streaming platform. Each userName can listen to multiple songs, and each song can be played by multiple users.
Your job is to process the data to:
 

1. Remove duplicate play records.
2. Track how many unique songs each userName has listened to.
3. Find the most popular song(s).
4. Output results in a structured format.
## Input
plays = [

("UserB", "Song1"),

("UserA", "Song2"),

("UserA", "Song1"),

("UserC", "Song3"),

]

Para manejar la data usaremos dos JSONs uno basado en los usuarios y otro en las canciones.

Para los usuarios guardaremos la canción con sus número de reproducciones por este usuario y el número de canciones únicas que ha escuchado el usuario.

En el otro caso (por canción) guardamos el número de reproducciones totales los usuarios que la han escuchado (con el número de reproducciones de cada usuario) y con estos datos determinamos la popularidad (tomamos el número de reproducciones totales y lo multiplicamos por el número de usuarios distintos que la han escuchado para evitar que un solo usuario suba mucho la popularidad de la canción pero que si afecte)

Después revisamos la información más general como la canción más popular (para esto creamos una función la cual nos ayuda a determinar la)

```python

import json

userData = {}
popularity = {}
generalData = {
   "mostPopular":("",0)
}
def getMostPoular(songName="", popularity=0):
    if songName!="" and popularity!=0 and generalData["mostPopular"][1] < popularity:
        generalData["mostPopular"] = (songName, popularity)
    return generalData["mostPopular"]

def processData(rawData=[]):
    for i in range(len(rawData)):
        userName = rawData[i][0]
        songName = rawData[i][1]

        # data by user
        userData.setdefault(userName, {"songs": {},"totalUniqueSongs": 0})
        userData[userName]["songs"][songName] = userData[userName]["songs"].get(songName, 0) + 1
        userData[userName]["totalUniqueSongs"] = len(userData[userName]["songs"])
        
        # data by song
        popularity.setdefault(songName,{"popularity":0, "plays":0 ,"listeners": {}})
        listenersLen = len(popularity[songName]["listeners"])
        if listenersLen==0: listenersLen=1 # evita hacer 0/0
        popularity[songName]["popularity"] = popularity[songName]["popularity"] / listenersLen
        popularity[songName]["popularity"] = popularity[songName]["popularity"] - popularity[songName]["plays"]
        popularity[songName]["listeners"][userName] = popularity[songName]["listeners"].get(userName, 0) + 1
        popularity[songName]["plays"] = popularity[songName]["plays"] + 1
        popularity[songName]["popularity"] = popularity[songName]["popularity"] + popularity[songName]["plays"]
        listenersLen = len(popularity[songName]["listeners"])
        popularity[songName]["popularity"] = popularity[songName]["popularity"] * listenersLen
        
        # general data 
        getMostPoular(songName, popularity[songName]["popularity"] )
```

Ahora creamos algunas funciones que nos ayudarán a mostrar la información.

```python
def getSong_JSON():
    return popularity
def getSong_List():
    data=[]
    for key, values in popularity.items():
        data.append({"Song": key,
                "popularity": values["popularity"],
                "plays": values["plays"],
                "listeners": (values["listeners"])
                })
    return data

def getUserSongs_JSON():
    return userData
def getUserSongs_List():
    data=[]
    for key, values in userData.items():
        data.append({"User": key,
                "totalUniqueSongs": values["totalUniqueSongs"],
                "songs": values["songs"],
                })
    return data
def getUserUniqueSongs():
    data=[]
    for key, values in userData.items():
        data.append({"User": key,"totalUniqueSongs": values["totalUniqueSongs"]})
    return data

def getUserSongs():
    data=[]
    for userName, values in userData.items():
        for songName, plays in values["songs"].items():
            data.append( (userName, songName) )
    return data

```

Ya con esto listo llamamos a la función para procesar la información y utilizamos las funciones requeridas para demostrar que se cumoplan los puntos.

```python

processData(plays)
print("1.- Remove duplicate play records: ")
print(getUserSongs())
print("2.- Track how many unique songs each userName has listened to: ")
print(getUserUniqueSongs())
print("3.- Find the most popular song(s): ")
print(getMostPoular())

print("4.- Output results in a structured format: ")
print("getSongs: ")
print(getSong_List())
print("getUserSongs: ")
print(getUserSongs_List())
```
