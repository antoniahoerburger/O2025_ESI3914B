# Lab 01 - Music Streaming Data Analysis

##  Description

You're given listening data from a music streaming platform. Each user can listen to multiple songs, and each song can be played by multiple users.

Your job is to process the data to:

##  Objectives

1. **Remove duplicate play records**
2. **Track how many unique songs each user has listened to**
3. **Find the most popular song(s)**
4. **Output results in a structured format**

---

##  Input Data

```python
plays = [
    ("UserA", "Song1"),
    ("UserB", "Song1"),
    ("UserA", "Song2"),
    ("UserA", "Song1"),
    ("UserC", "Song3"),
    ("UserB", "Song2"),
    ("UserD", "Song1"),
    ("UserC", "Song1"),
    ("UserD", "Song3"),
]
```

---

````python
# 1. ALGORITMO PARA REMOVER DUPLICADOS
plays = [
    ("UserA", "Song1"), ("UserB", "Song1"), ("UserA", "Song2"), 
    ("UserA", "Song1"), ("UserC", "Song3"), ("UserB", "Song2"), 
    ("UserD", "Song1"), ("UserC", "Song1"), ("UserD", "Song3")
]


def remove_duplicates(plays_list):
    # Convertir a set elimina automáticamente los duplicados
    unique_plays = list(set(plays_list))
    return unique_plays

unique_plays = remove_duplicates(plays)
````

````python
# 2. ALGORITMO PARA CONTAR CANCIONES ÚNICAS POR USUARIO
def count_unique_songs_per_user(plays_list):
    user_songs = {}
    
    for user, song in plays_list:
        if user not in user_songs:
            user_songs[user] = set()
        user_songs[user].add(song)
    user_song_counts = {user: len(songs) for user, songs in user_songs.items()}
    return user_song_counts, user_songs

user_counts, user_song_sets = count_unique_songs_per_user(unique_plays)
````

````python
# 3. ALGORITMO PARA ENCONTRAR LAS CANCIONES MÁS POPULARES
def find_most_popular_songs(plays_list):
    song_counts = {}
    for user, song in plays_list:
        if song not in song_counts:
            song_counts[song] = 0
        song_counts[song] += 1
    
    max_plays = max(song_counts.values()) if song_counts else 0
    
    most_popular = [song for song, count in song_counts.items() if count == max_plays]
    
    return song_counts, most_popular, max_plays

song_counts, most_popular, max_plays = find_most_popular_songs(unique_plays)
````

````python
# 4. FORMATO ESTRUCTURADO DE RESULTADOS
def generate_structured_output(unique_plays, user_counts, song_counts, most_popular):
    result = {
        "summary": {
            "total_unique_plays": len(unique_plays),
            "total_users": len(user_counts),
            "total_unique_songs": len(song_counts)
        },
        "user_analysis": user_counts,
        "song_popularity": song_counts,
        "most_popular_songs": most_popular
    }
    
    return result

structured_results = generate_structured_output(unique_plays, user_counts, song_counts, most_popular)

print("4. RESULTADOS ESTRUCTURADOS:")
print("\n RESUMEN GENERAL:")
print(f"   • Total de reproducciones únicas: {structured_results['summary']['total_unique_plays']}")
print(f"   • Total de usuarios: {structured_results['summary']['total_users']}")
print(f"   • Total de canciones únicas: {structured_results['summary']['total_unique_songs']}")

print("\n ANÁLISIS POR USUARIO:")
for user, count in sorted(structured_results['user_analysis'].items()):
    print(f"   • {user}: {count} canciones únicas")

print("\n RANKING DE CANCIONES:")
# Ordenar canciones por popularidad (descendente)
sorted_songs = sorted(structured_results['song_popularity'].items(), 
                     key=lambda x: x[1], reverse=True)
for i, (song, count) in enumerate(sorted_songs, 1):
    print(f"   {i}. {song}: {count} reproducciones")

print(f"\n CANCIÓN(ES) MÁS POPULAR(ES): {', '.join(structured_results['most_popular_songs'])}")
````
